```python
df['veh_type'].fillna(df['veh_type'].mode()[0], inplace=True)
df['priority'].fillna(df['priority'].mode()[0], inplace=True)
df['corridor'].fillna(df['corridor'].mode()[0], inplace=True)
df['endlatitude'].fillna(df['endlatitude'].mode()[0], inplace=True)
df['endlongitude'].fillna(df['endlongitude'].mode()[0], inplace=True)
```

```python
df = pd.get_dummies(df, columns=['event_type', 'event_cause','requires_road_closure','status','authenticated','veh_type','corridor','priority','police_station'], drop_first=True)
```

```python
df.head()
```

```python
df['start_datetime'] = pd.to_datetime(df['start_datetime'])
```

```python
df['modified_datetime'] = pd.to_datetime(df['modified_datetime'])
df['created_date'] = pd.to_datetime(df['created_date'])
```

```python
df['start_datetime'] = pd.to_datetime(
    df['start_datetime'],
    format='mixed'
)
df['modified_datetime'] = pd.to_datetime(
    df['modified_datetime'],
    format='mixed'
)

df['created_date'] = pd.to_datetime(
    df['created_date'],
    format='mixed'
)
```

```python
# start_datetime features
df['start_year'] = df['start_datetime'].dt.year
df['start_month'] = df['start_datetime'].dt.month
df['start_day'] = df['start_datetime'].dt.day
df['start_hour'] = df['start_datetime'].dt.hour
df['start_dayofweek'] = df['start_datetime'].dt.dayofweek

# modified_datetime features
df['modified_year'] = df['modified_datetime'].dt.year
df['modified_month'] = df['modified_datetime'].dt.month
df['modified_day'] = df['modified_datetime'].dt.day
df['modified_hour'] = df['modified_datetime'].dt.hour

# created_date features
df['created_year'] = df['created_date'].dt.year
df['created_month'] = df['created_date'].dt.month
df['created_day'] = df['created_date'].dt.day
df['created_dayofweek'] = df['created_date'].dt.dayofweek
```

```python
df['modify_delay_hours'] = (
    df['modified_datetime'] - df['start_datetime']
).dt.total_seconds() / 3600

df['creation_to_start_hours'] = (
    df['start_datetime'] - df['created_date']
).dt.total_seconds() / 3600
```

```python
df.drop(
    columns=['start_datetime', 'modified_datetime', 'created_date'],
    inplace=True
)
```

```python
df.drop('address', axis=1, inplace=True)
```
